```python
import numpy as np
from scipy.io import wavfile
import pandas as pd
import os
import matplotlib.pyplot as plt
from datetime import timedelta, datetime

os.mkdir("time_series")
for file in os.listdir(".\\signals"):
    
    sampling_rate, signal = wavfile.read("signals\\"+file)

    if len(signal.shape) == 2:
        signal = signal.mean(axis=1)  

    time = np.linspace(0, len(signal) / sampling_rate, num=len(signal))

    time_series_data = pd.DataFrame({'time': time, 'amplitude': signal})

    time_series_data.to_csv(f"time_series\\{file.split(".")[0]}.csv", index=False)

    plt.figure(figsize=(10, 6))
    plt.plot(time, signal)
    plt.title(f"{file.split(".")[0]}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()
```
